# Kernel check (VS Code + Jupyter)

This notebook helps debug kernel startup issues on Windows.

- The first code cell runs under **Python** and lists installed Jupyter kernelspecs (works even if the R kernel won’t start).
- The second code cell is **R** for a quick sanity check once you can start an R kernel.

If you opened this repo as a GitHub virtual filesystem workspace (`vscode-vfs://...`), prefer cloning locally for kernel reliability.

```python
import json
import os
import sys
from jupyter_client.kernelspec import KernelSpecManager

print('Python:', sys.version)
ksm = KernelSpecManager()
specs = ksm.find_kernel_specs()  # name -> path
print(f'Found {len(specs)} kernelspec(s).')
for name, path in sorted(specs.items()):
    print(f'- {name}: {path}')

# Optional: inspect one kernelspec in detail
kernel_name = 'wilsontew-r'
if kernel_name in specs:
    kernel_json = os.path.join(specs[kernel_name], 'kernel.json')
    print('\n---')
    print(f'{kernel_name} kernel.json path: {kernel_json}')
    try:
        with open(kernel_json, 'r', encoding='utf-8') as f:
            print(json.dumps(json.load(f), indent=2))
    except Exception as e:
        print('Could not read kernel.json:', e)
else:
    print('\nNote: wilsontew-r kernelspec not found (yet).')
```
