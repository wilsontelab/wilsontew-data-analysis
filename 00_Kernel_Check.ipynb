# Kernel check / future-proof notebook

This notebook is a lightweight, keep-forever utility to help you maintain a clean `.ipynb` workflow in this repo.

Jupyter notebooks typically run **one kernel per notebook**. To see what kernels are available, run the cell below.

If you need to add kernels later, see `KERNELS.md` (and `environment.yml` for a reproducible conda setup).

```r
# List available Jupyter kernels from within R
cat('R:', R.version.string, '\n\n')

jupyter <- Sys.which('jupyter')
out <- character(0)

if (jupyter == '') {
  cat('jupyter executable not found on PATH.\n')
  cat('If you are using conda, activate the repo environment (see environment.yml) and try again.\n')
} else {
  args <- c('kernelspec', 'list')
  cat('Running:', shQuote(jupyter), paste(args, collapse = ' '), '\n')
  cat(strrep('-', 60), '\n')

  out <- tryCatch(
    system2(jupyter, args, stdout = TRUE, stderr = TRUE),
    error = function(e) structure(conditionMessage(e), class = 'kernel_check_error')
  )

  if (inherits(out, 'kernel_check_error')) {
    cat('Error running jupyter:', as.character(out), '\n')
  } else {
    cat(paste(out, collapse = '\n'), '\n')
  }

  wanted <- list(
    python = c('python', 'python3'),
    r = c('ir', 'irkernel', 'wilsontew-r'),
    java = c('java', 'ijava'),
    `c++` = c('xeus-cling', 'cling'),
    dotnet = c('dotnet'),
    powershell = c('powershell')
  )

  cat(strrep('-', 60), '\n')
  cat('Heuristic check for common kernel names:\n')
  lower <- tolower(paste(out, collapse = '\n'))
  for (label in names(wanted)) {
    tokens <- wanted[[label]]
    hit_idx <- which(vapply(tokens, function(t) grepl(t, lower, fixed = TRUE), logical(1)))
    hit <- if (length(hit_idx)) tokens[[hit_idx[[1]]]] else NA_character_
    if (!is.na(hit)) {
      cat(sprintf('  %-12s -> FOUND (%s)\n', label, hit))
    } else {
      cat(sprintf('  %-12s -> not found\n', label))
    }
  }
}
```

## Notes on the languages you asked about

- **R**: use **IRkernel** (see `KERNELS.md`).
- **Python**: included by default via `ipykernel`.
- **Java**: use **IJava** (requires a JDK).
- **C++**: use **xeus-cling**.
- **VBA**: there is no broadly supported Jupyter VBA kernel. If you need Office automation, you can usually drive VBA/Excel from Python (e.g., `xlwings`, `pywin32`) or PowerShell instead.
